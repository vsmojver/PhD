```python
url="C:\\Users\\vsmojver\\Google Drive\\Doktorski\\Datasets\\Battery\\battery_backward_sve.csv"
#url="C:\\Users\\Vladimirko\\Google Drive\\Doktorski\\Datasets\\AdMan\\Splits\\AdMan_backward_11000.csv"
#url="C:\\Users\\vsmojver\\Google Drive\\Doktorski\\Datasets\\\\NFC\\nfc_backward.csv"
```

```python
#pazi koji je delimiter ";" ili ","
#prvu liniju koristis kad treba preskocit redove, drugu kad ucitavas cijeli dset
#skiprows=range(1,)
#nrows
A= pd.read_csv(url,index_col=1, delimiter=',',  low_memory=False, skiprows=range(1,2000), nrows=2000)
#A= pd.read_csv(url,index_col=1, delimiter=',',  low_memory=False)
A=A.drop(A.columns[0], axis=1)
A=A.dropna(axis=1, how='all') #brise columne gdje nema citata
A=A.dropna(axis=0, how='all') #brise redove gdje nema citata
#A=A.fillna('nula')
A
```

```python
n=1                             #broj chunkova
list_A=[A[i:i+n].dropna(axis=1) for i in range(0, A.shape[0], n)]
len(list_A)
```

```python
print(list_A[8])
```

```python
edge=[[] for _ in range(len(list_A))]
```

```python
for a in range(0,len(list_A)):
    print(a)
    for index, rows in list_A[a].iterrows():
        for comb in (it.combinations(rows,2)):
            edge[a].append(comb)  
```

```python
edge_spojeno=[j for i in edge for j in i]
```

```python
edge_pd=pd.DataFrame(edge_spojeno).dropna()
edge_pd
```

```python
edge_pd.to_csv("battery_edge_2000-4000.csv", header=False, index=False)
```
